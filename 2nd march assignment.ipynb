## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data

plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.


A scatter plot is a set of points plotted on a horizontal and vertical axes. Scatter plots are important in statistics because they can show the extent of correlation, if any, between the values of observed quantities or phenomena (called variables).

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,c="r")
plt.xlabel("This is the x label of the scattered plot")
plt.ylabel("This is the y label of the scattered plot")
plt.title("X vs Y plot of the scattered data")
```

subplot divides the current figure into rectangular panes that are numbered row-wise. Each pane contains an axes. Subsequent plots are output to the current pane. subplot(m,n,p) creates an axes in the p -th pane of a figure divided into an m -by- n matrix of rectangular panes.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.figure(constrained_layout=True)
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.xlabel("This is my x1 axis")
plt.ylabel("This is my y1 axis")
plt.title("This is my x1 vs y1 graph")
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.xlabel("This is my x2 axis")
plt.ylabel("This is my y2 axis")
plt.title("This is my x2 vs y2 graph")
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.xlabel("This is my x3 axis")
plt.ylabel("This is my y3 axis")
plt.title("This is my x3 vs y3 graph")
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.xlabel("This is my x4 axis")
plt.ylabel("This is my y4 axis")
plt.title("This is my x4 vs y4 graph")
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])



Bar plots are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the category of the data. 

Bar charts should be used when we are showing segments of information. Vertical bar charts are useful to compare different categorical or discrete variables, such as age groups, classes, schools, etc., as long as there are not too many categories to compare. They are also very useful for time series data.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(12,8),constrained_layout=True)
plt.subplot(2,2,1)
plt.bar(company,profit)
plt.xlabel("Company names")
plt.ylabel("Profit by the company in crores")
plt.title("Graph of company vs profit earned")
plt.subplot(2,2,3)
plt.barh(company,profit)
plt.ylabel("Company names")
plt.xlabel("Profit by the company in crores")
plt.title("Graph of company vs profit earned")
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot is a special type of diagram that shows the quartiles in a box and the line extending from the lowest to the highest value.

Box plots are used to show distributions of numeric data values, especially when we want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
fig = plt.figure(figsize =(10, 7))

bp = plt.boxplot(data)
```
